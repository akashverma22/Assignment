### Dimensionality Reduction with PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
df1=data.copy()
```

```python
std_sca=StandardScaler()
```

```python
features=pd.DataFrame(std_sca.fit_transform(df1),columns=df1.columns)
features
```

```python
## dataset has been standardized. Now we will apply PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=14)  ## we define PCA
```

```python
pca.fit(features)
```

```python
features_new=pca.transform(features)
```

```python
print(features_new.shape)
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
## PCA1 contains 40% information of the features, PCA2 contains 18% information, PCA3 contains 10% information and so on
```

```python
## We will check the explained variance by the help of a graph
```

```python
plt.plot(range(1, len(explained_variance) + 1),explained_variance*100,marker='o')
plt.title('Explained Variance Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance in percentage')
plt.grid(True)
plt.show();
```

```python
## graph shows first six PCAs are sufficient to explain the data and contain almost 90% of the information.
## Hence optimum number of PCA should be 6.
```

```python
## Now we transform our features to contain six columns
```

```python
pca=PCA(n_components=6)
```

```python
pca.fit(features)
```

```python
features_new=pca.transform(features)
```

```python
print(features_new.shape)
```

```python
pca.explained_variance_ratio_
```

```python
## This completes data transformation using PCA. We will preserve this transformed data for further use
```

### Clustering with Original Data

##### K-Means Clustering

```python
features
```

```python
## we have already checked there are no duplicates or null values hence we start building model
```

##### K-Means clustering

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
## for Elbow method first we calculate inertia for 10 clusters
inertia=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
print(inertia)
```

```python
## plotting the graph between number of clusters and inertia to find the optimum number of clusters
plt.plot(range(1,11),inertia)
plt.xlabel('No_of_clusters')
plt.ylabel('Inertia')
plt.title('Elbow_Method',color='red',size=20)
plt.show();
```

```python
## From the graph it is evident that the optimum number of clusters is 3
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
cluster=kmeans.fit(features)
```

```python
cluster.labels_
```

```python
np.unique(cluster.labels_)
```

```python
## centroids for K=3 clusters with 13 variables
cluster.cluster_centers_
```

```python
## we add the cluster column to a copy of original dataset
df1=data.copy()
df1['No_of_clusters']=cluster.labels_
df1
```

```python
## grouping dataset cluster-wise on the basis of mean
df1.groupby('No_of_clusters').agg(['mean']).reset_index()
```

```python
## visualizing clusters on the basis of Alcohol column
plt.figure(figsize=(10,6))
plt.scatter(df1['No_of_clusters'],df1['Alcohol'],c=cluster.labels_)
plt.show();
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(features,cluster.labels_)
```

```python
# Accuracy as per KMeans is 31 percent
```

### Comparison and Analysis

```python
# Comparison of Clustering Results from Original Dataset vs. PCA-transformed Data:

# Original Dataset: Clustering directly on the original dataset uses all the available features, which could lead to capturing more complex patterns but might also be influenced by noisy or irrelevant features. The clusters formed in the original dataset might reflect some variability that is not fully meaningful due to high dimensionality.
# PCA-transformed Data: PCA reduces the dimensionality by capturing the most important variance in fewer components. When clustering is performed on the PCA-transformed data, the algorithm operates on fewer dimensions that are assumed to carry the most relevant information.

# Comparison:
# The clustering results from the PCA-transformed data may show clearer and more distinct clusters if the dimensionality reduction successfully eliminates noise and irrelevant features.
# On the other hand, the original dataset may form slightly overlapping clusters because some variables may not contribute effectively to separation between clusters.

# Similarities or Differences Observed:
# Similarities: If the PCA components have captured most of the essential variance, the overall structure of the clusters in both datasets might be similar. For example, the same number of clusters and their general positioning in feature space may be comparable.
# Differences: The cluster boundaries and intra-cluster compactness may differ. PCA-transformed clusters might be more tightly grouped, while those in the original dataset might show greater spread due to the influence of non-essential features.

# Impact of Dimensionality Reduction on Clustering Performance:
# PCA helps in removing noise and reducing the complexity of the feature space. By focusing on the most significant components, it simplifies the clustering task for algorithms like K-means or DBSCAN, which can more easily identify distinct patterns.
# However, this reduction also involves the loss of some information. If important variability is captured by components that were discarded during PCA, clustering performance might suffer.
# Metrics like the silhouette score or Daviesâ€“Bouldin index could indicate whether the clustering on the PCA-reduced data improves cluster compactness and separation, or if it negatively impacts the result due to oversimplification.

# Trade-offs Between PCA and Clustering on Original Data:

# Advantages of PCA + Clustering:
# Faster computation due to reduced feature space.
# Clearer, more distinct clusters if noise or redundant features are present in the original data.
# Reduced risk of overfitting due to fewer dimensions.

# Disadvantages:
# Loss of some information, which might lead to the omission of relevant patterns.
# If the dataset already has strong inherent structure, PCA might not provide significant improvement, and clustering on the original dataset may be more accurate.

# Advantages of Clustering on Original Data:
# Retains all information, allowing the clustering algorithm to consider all potential patterns.
# May lead to better results if the dataset has low noise and high interpretability in higher dimensions.

# Disadvantages:
# Slower computational speed, especially with high-dimensional data.
# The algorithm may be misled by noise or irrelevant features, leading to poorer clustering performance.
```

### Conclusion and Insights

```python
# Key Findings and Insights:

# Dimensionality reduction through PCA can significantly improve clustering performance by simplifying the feature space and removing noise. However, it comes with the trade-off of losing some information.
# The clustering results can vary depending on whether PCA is applied, and this highlights the importance of balancing dimensionality reduction with retaining key patterns in the data.
# Depending on the datasetâ€™s structure, PCA may enhance or reduce the quality of clusters, which should be evaluated on a case-by-case basis.

# Practical Implications of Using PCA and Clustering:

# For High-Dimensional Data: When dealing with high-dimensional data, PCA is useful to reduce complexity, improve computational efficiency, and potentially lead to better clustering outcomes. In practice, PCA should be considered when the dataset has many redundant or highly correlated features.
# For Low-Dimensional Data: In situations where the dataset is already low-dimensional or has strong inherent structure, PCA may not be necessary, and clustering on the original data might be preferable.
# Model Evaluation: Itâ€™s essential to evaluate clustering results using metrics like silhouette scores before and after PCA to ensure that the reduction in dimensions doesnâ€™t come at the cost of important data patterns.

# Recommendations for Using PCA and Clustering:

# When to Use PCA:
# Use PCA before clustering when working with high-dimensional data to simplify the feature space and remove noise.
# Apply PCA when the clustering algorithm struggles to form distinct clusters in the original feature space due to overlapping data or when computational efficiency is important.

# When to Cluster on Original Data:
# Skip PCA if the data is already low-dimensional or if every feature is considered crucial to the problem at hand.
# Consider clustering directly if your dataset has low noise, a small number of features, or meaningful relationships among all variables.
```
